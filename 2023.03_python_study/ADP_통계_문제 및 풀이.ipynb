#### 2. 다음은 암컷 원숭이의 몸무게를 조사한 자료이다(단위 kg). 원숭이의 몸무게는 분산이 1인 정규분포를 따른다고 할 수 있다고 한다. 암컷 원숭이의 평균 몸무게가 8.5kg라고 할 수 있는지 검정하시오.
8.3, 9.5, 9.6, 8.75, 8.4, 9.1, 8.15, 8.8

```python
x = [8.3, 9.5, 9.6, 8.75, 8.4, 9.1, 8.15, 8.8]
```

```python
stats.ttest_1samp(x, popmean = 8.5)
```

```python
import numpy as np
import scipy.stats as stats

data = np.array([8.3, 9.5, 9.6, 8.75, 8.4, 9.1, 8.15, 8.8])
null_hypothesis_mean = 8.5
alpha = 0.05

sample_mean = np.mean(data)
sample_size = len(data)
standard_deviation = 1
standard_error = standard_deviation / np.sqrt(sample_size)

z_statistic = (sample_mean - null_hypothesis_mean) / standard_error
p_value = stats.t.sf(np.abs(z_statistic), df=sample_size-1) * 2

print("z-statistic:", z_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("귀무 가설 기각: 암컷 원숭이의 평균 몸무게는 8.5kg이 아니다.")
else:
    print("귀무 가설 채택: 암컷 원숭이의 평균 몸무게가 8.5kg이라고 할 수 있다.")
```

#### 4/17 (월) : 모비율의 신뢰구간

- Q. 특정 제품 A의 선호와 비선호를 알아보기 위해 1,000명을 대상으로 조사한 결과 A를 선호하는 사람들이 520명임을 알게 되었다. 제품 A의 선호율에 대해 물음에 답하시오.
    - (1) 신뢰수준 90%의 신뢰구간을 구하라.
    - (2) 신뢰수준 95%의 신뢰구간을 구하라
    - (3) 만일 4,000명을 대상으로 조사하였을 때, A를 2,080명이 선호하였다면 신뢰수준 95%에서 A 선호율에 대한 신뢰구간을 구하고 (2)의 결과와 비교하라.

```python
import math
from scipy.stats import norm

# 90%와 95%의 z-값을 구합니다.
z_90 = norm.ppf(0.95)
z_95 = norm.ppf(0.975)

def confidence_interval(sample_size, num_success, z):
    p = num_success / sample_size
    margin_of_error = z * math.sqrt(p * (1 - p) / sample_size)
    lower = p - margin_of_error
    upper = p + margin_of_error
    return lower, upper

# (1) 90% 신뢰수준의 신뢰구간
lower_90, upper_90 = confidence_interval(1000, 520, z_90)
print(f"90% 신뢰수준의 신뢰구간: ({lower_90:.4f}, {upper_90:.4f})")

# (2) 95% 신뢰수준의 신뢰구간
lower_95, upper_95 = confidence_interval(1000, 520, z_95)
print(f"95% 신뢰수준의 신뢰구간: ({lower_95:.4f}, {upper_95:.4f})")

# (3) 4,000명을 대상으로 조사한 경우
lower_95_large_sample, upper_95_large_sample = confidence_interval(4000, 2080, z_95)
print(f"신뢰수준 95%에서 4,000명을 대상으로 조사한 경우 A 선호율에 대한 신뢰구간: ({lower_95_large_sample:.4f}, {upper_95_large_sample:.4f})")
```

```python
# 모비율 p의 신뢰구간

# 신뢰수준 90%일 때
import numpy as np

p=0.52
n=1000
a1 = p-np.sqrt((p*(1-p))/n)*1.645
a2 = p+np.sqrt((p*(1-p))/n)*1.645

print(round(a1*100,3),round(a2*100,3))

# 신뢰수준 95%일 때
p=0.52
n=1000
b1 = p-np.sqrt((p*(1-p))/n)*1.96
b2 = p+np.sqrt((p*(1-p))/n)*1.96

print(round(b1*100,3),round(b2*100,3))

# n=4000일 때
p=0.52
n=4000
c1 = p-np.sqrt((p*(1-p))/n)*1.96
c2 = p+np.sqrt((p*(1-p))/n)*1.96

print(round(c1*100,3),round(c2*100,3)) # (2)와 비교하여 신뢰구간의 폭이 더 작아졌다. 표본이 많을수록 더 정확한 결과를 얻을 수 있음을 알 수 있다.
```

    - {2610, 2750, 2420, 2510, 2540, 2490, 2680}
    - (1) 이 금융기관의 고객 대기 시간 평균이 2,500초를 넘는다고 할 수 있는가를 검정하기 위한 가설을 설정하라
    - (2) 유의수준 0.05에서 가설검정하라
    - (3) 유의수준을 0.1로 할 경우, (2)의 결과가 상이한가?
    - (4) p값을 구하여 (2)와 (3)의 결과와 비교하고 평가하라.

```python
x = [2610, 2750, 2420, 2510, 2540, 2490, 2680]
from scipy.stats import shapiro
import scipy.stats as stats
```

```python
shpro = shapiro(x)
print('H0 : 고객 대기시간 측정 결과는 정규성을 가진다.')
print('H1 : 고객 대기시간 측정 결과는 정규성을 가지지 않는다.')
print(shpro)
if shpro.pvalue < 0.05:
    print("결과 : 귀무 가설 기각 - 고객 대기시간 측정 결과는 정규성을 가지지 않는다.")
else:
    print("결과 : 귀무 가설 채택 - 고객 대기시간 측정 결과는 정규성을 가진다")
```

```python
# 가설 설정
print('H0 : 고객 대기시간의 평균이 2500초 이하이다.')
print('H1 : 고객 대기시간의 평균이 2500초 초과이다.')
```

```python
tt1 = stats.ttest_1samp(x, popmean = 2500, alternative = 'greater')
print(tt1)

if tt1.pvalue < 0.05:
    print("결과 : 귀무 가설 기각 - 고객 대기시간의 평균이 2500초 초과이다.")
else:
    print("결과 : 귀무 가설 채택 - 고객 대기시간의 평균이 2500초 이하이다.")
```

```python
tt2 = stats.ttest_1samp(x, popmean = 2500, alternative = 'greater')
print(tt2)
if tt2.pvalue < 0.1:
    print("결과 : 귀무 가설 기각 - 고객 대기시간의 평균이 2500초 초과이다.")
else:
    print("결과 : 귀무 가설 채택 - 고객 대기시간의 평균이 2500초 이하이다.")
```

```python
from scipy.stats import chi2_contingency

result = chi2_contingency(df)
print('H0 : 근무시간대에 따른 불량률이 독립이다.(차이가 없다.)')
print('H1 : 근무시간대에 따른 불량률이 독립이 아니다.(차이가 있다.)')
print(result)
if result[1] < 0.05:
    print("결과 : 귀무 가설 기각 - 근무시간대에 따른 불량률이 독립이 아니다.(차이가 있다.)")
else:
    print("결과 : 귀무 가설 채택 - 근무시간대에 따른 불량률이 독립이다.(차이가 없다.)")
```

- H0: 4개의 교육훈련 기법 간 차이가 없다.
- H1: 4개의 교육훈련 기법 간 차이가 있다.

```python
df_stack = pd.concat([pd.DataFrame({'value':a, 'label':'a'}), 
                      pd.DataFrame({'value':b, 'label':'b'}),
                     pd.DataFrame({'value':c, 'label':'c'}),
                      pd.DataFrame({'value':d, 'label':'d'})], axis = 0)
df_stack.head()
```

```python
# 사후검정
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(data = df_stack.value,groups = df_stack.label)
tuekeyhsd = mc.tukeyhsd(alpha = 0.05)
fig = tuekeyhsd.plot_simultaneous()
```

```python
tuekeyhsd.summary()
```

- H1 : group1과 group2의 평균은 다르다.

지역 = ['A', 'A', 'A', 'B', 'A', 'B', 'B', 'B', 'B']    
순위 = [1, 2, 3, 4, 5, 6, 7, 8, 9]    
지역 A와 B에 생산된 라이터 돌의 강도를 비교하는 가설을 세우고 검정하시오.

#### Q. 랜덤하게 선택된 10명의 자연계열 신입생에게 심리검사를 실시한 결과 책임감에 대한 점수가 다음과 같이 나타났다.
28.5, 25.2, 28.7, 41, 29.1, 32.3, 37.7, 39.9, 26.8, 28.8    
평균점수가 30점인지를 유의수준 0.05로 검정해보자.     


```python
data = [28.5, 25.2, 28.7, 41, 29.1, 32.3, 37.7, 39.9, 26.8, 28.8]
```

```python
stats.shapiro(data)
```

H0: 정규분포를 따른다.    
H1: 정규분포를 따르지 않는다.    
유의확률이 0.074로 유의수준 0.05보다 크므로 귀무가설을 채택한다.
즉, 정규분포를 따른다. 

```python
stats.ttest_1samp(data, 30)
```

H0: 책임감 점수의 평균 점수는 30점이다.    
H1: 책임감 점수의 평균 점수는 30점이 아니다.    
유의확률이 0.342로 유의수준 0.05보다 크므로 귀무가설을 채택한다. 책임감 점수의 평균 점수는 30점이다.

```python
# 비모수인 경우
stan_value = list(np.repeat(30, len(data)))
stats.wilcoxon(stan_value, data)
```

H1: 책임감 점수의 평균 점수는 30점이 아니다.
