## 제9절. 나이브베이즈(Naive Bayes) 분류


- 나이브베이즈 : 데이터가 각 클래스에 속할 특징 확률을 계산하는 조건부 확률 기반의 분류 방법
    - 나이브(Naive) '단순한' 이라는 의미로 예측한 특징을 상호독립적이라는 가정 하에 확률 계산을 단순화하고 모든 변수가 동등하다는 특징에서 비롯됨.
    - 베이즈 : 베이지안 통계를 기반으로 입력특징이 클래스 전체의 확률 분포 대비 특정 클래스에 속할 확률을 정리하였다는 점에 비롯됨.


### 베이즈 정리(Bayes'Theorem)
- 두 확률 변수의 사전 확률과 사후 확률 사이의 관계를 나타내는 정리
- 목적 : 사건 A, B가 있을 때, 사후 확률 P(A|B) 구하기 (사건 B가 일어난 것을 전제로 한 사건 A의 조건부 확률을 구하는 것)
- 알고 있는 것 : P(B) - 관찰값(Evidence) <br>
                  P(A) - 사전 확률(prior)  <br>
                  P(B|A) - 우도(likehood)

- 사전에 A에 대한 정보를 바탕으로 P(A)와 P(B|A)를 알고 있다면, 이를 활용해 P(A|B)를 추론할 수 있게 해주는 것,
- 새로운 증거 P(B)에 기반하여 과거의 정보를 향상시키거나 개선해줄 수 있는 이론

- 조건부 확률

- 곱셈 공식 <br>
P(A∩B) = P(A|B)P(B) = P(B∩A) = P(B|A)P(A) <br>
A와 B가 독립시행일 경우 P(A∩B) = P(A)*P(B)

- 베이즈 정리
    - 전제조건 <br>
      1. 표본공간 S는 서로소인 A1, A2, ..., An의 합집합 <br>
      2. B는 S 위에서 정의된 사건, P(A)!=0


#### 나이브베이즈 분류
- 나이브베이즈 분류에서의 베이즈 정리
    1. 먼저, 주어진 데이터를 이용해 클래스별 사전 확률 P(C)와 각 특징이 주어졌을 때 클래스의 확률 P(x_i|C)를 계산
    2. 새로운 데이터 포인트가 주어졌을 때, 베이즈 정리를 사용하여 각 클래스에 대한 사후 확률 P(C|x)를 구함
    3. 구한 사후 확률 중 가장 큰 값을 가진 클래스를 선택하여 새로운 데이터 포인트의 클래스를 예측
- 
- 장단점
    - 장점 : 간단하고 빠름 / 노이즈 및 결측치에 강함
    - 단점 : 모든 특징이 동등하게 중요하며 독립이라고 가정하기 때문에 가정이 잘못된 경우가 종종 있음.
- 데이터가 많지 않을 때, 미래 예측일 때 나이브베이즈 분류 기법을 추천

### 1. 가우시안
#### (1) scikit-learn의 GaussianNB

- 코드 실습

```python
import pandas as pd
sky = pd.read_csv("/kaggle/input/adp-book-data/Skyserver.csv")
sky.info()
```

```python
sky['class'].unique()
```

```python
import seaborn as sns
sns.pairplot(hue='class', data =sky[['z', 'run', 'i', 'class']])
```

```python
import numpy as np
features = list(sky.columns)
features.remove('class')
X = sky[features]
y = sky['class']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state =1, stratify =y)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
pred = gnb.fit(x_train, y_train).predict(x_test)
print("Accuracy Score : ", gnb.score(x_test, y_test))
```

```python
gnb.predict_proba(x_test)[[0 , 13 , 68 ]]
```

```python
gnb.predict(x_test)[[0 , 13 , 68 ]]
```

```python
from sklearn.metrics import classification_report
pred=gnb.predict(x_test)
print(classification_report(y_test, pred))
```

## 2. 베르누이
### (1) scikit-learn의 BernoulliNB

- 코드 실습

```python
import pandas as pd
spam = pd.read_csv('/kaggle/input/adp-book-data/spam.csv', encoding ='utf-8')
spam.info()
```

```python
spam.isna().sum()
```

```python
spam=spam[['v1', 'v2']]
spam
```

```python

spam['v1'].unique()
```

```python
import numpy as np
spam['label'] = np.where(spam['v1']=='spam', 1, 0)
spam
```

```python
X= spam['v2']
y = spam['label']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state =1, stratify =y)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(binary =True)
x_traincv = cv.fit_transform(x_train)
x_traincv.shape
```

```python
encoded_input = x_traincv.toarray()
encoded_input
```

```python
print(cv.inverse_transform(encoded_input[[0]]))
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_traincv, y_train)
```

```python
x_testcv = cv.transform(x_test)
pred = bnb.predict(x_testcv)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print("Accuracy Score : ", acc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

### 2. 다항 나이브베이즈
#### (1) scikit-learn의 MultinomialNB

- 코드 실습

```python
from keras.datasets import imdb
(X_train, y_train), (X_test, y_test) = imdb.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
import pandas as pd

word_to_index = imdb.get_word_index()
index_to_word = {}

for key, value in word_to_index.items():
    index_to_word[value+3] = key
for index, token in enumerate(("<pad>", "<sos>", "<unk>")):
    index_to_word[index] = token

train_reviews = []
for X in X_train:
    tmp =' '.join([index_to_word[index] for index in X])
    train_reviews.append(tmp)

test_reviews = []
for X in X_test:
    tmp =' '.join([index_to_word[index] for index in X])
    test_reviews.append(tmp)

train = pd.concat([pd.DataFrame(train_reviews), pd.DataFrame(y_train)], axis =1)
train.columns = ['reviews', 'label']
train['reviews'] = train['reviews'].str[6:]

test = pd.concat([pd.DataFrame(test_reviews), pd.DataFrame(y_test)], axis =1)
test.columns = ['reviews', 'label']
train['reviews'] = train['reviews'].str[6:]

print("<<<<<<<<< Train Dataset for MNB >>>>>>>>>", train)
print("<<<<<<<<< Test Dataset for MNB >>>>>>>>>", test)
```

```python
x_train, x_test = train['reviews'].values, test['reviews'].values
y_train, y_test = train['label'].values, test['label'].values
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=False)
x_traincv = cv.fit_transform(x_train)
x_traincv.shape
```

```python
x_traincv[0]
```

```python
print (cv.inverse_transform(x_traincv)[0])
```

```python
print (cv.get_feature_names()[-10 :])
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(x_traincv, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
x_testcv = cv.transform(x_test)
pred = mnb.predict(x_testcv)
acc = accuracy_score(y_test, pred)
print("Accuracy Score : ", acc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
